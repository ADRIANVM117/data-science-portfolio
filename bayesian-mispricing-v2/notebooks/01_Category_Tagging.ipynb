```python
import pandas as pd
import numpy as np
import re
from pathlib import Path

pd.set_option("display.max_columns", None)
df_market = pd.read_csv( "../data/processed/market_dataset.csv")
df = df_market.copy()
df["category"].value_counts(dropna=False)


def classify_market(question):

    q = str(question).lower()

    if any(x in q for x in [
        "fed",
        "interest rates",
        "rate cut",
        "inflation",
        "gdp",
        "usd/try",
        "debt ceiling"
    ]):
        return "MONETARY_MACRO"

    if any(x in q for x in [
        "biden",
        "desantis",
        "election",
        "house control",
        "congress",
        "guatemala",
        "ukraine",
        "iran",
        "russian",
        "ceasefire",
        "prigozhin"
    ]):
        return "POLITICS_GEO"

    return "EVENTS_OTHER"
```

```python
df["research_group"] = df["question"].apply(classify_market)
df["research_group"].value_counts()
df.to_csv("../data/processed/market_dataset_tagged.csv",index=False)
```
